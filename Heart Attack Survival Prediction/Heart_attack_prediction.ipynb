# Model Training

```python
train_data = train_data.sample(frac=1).reset_index(drop = True)
```

```python
cols_model = ['age','sex','cp','trtbps','chol','fbs','restecg','thalachh','exng','oldpeak','slp','caa','thall']
train_X = train_data[cols_model]
train_Y = train_data["output"]
```

```python
train_X = train_X.values
train_Y = train_Y.values
#converting the pandas dataframe into a numpy nd array
```

```python
split = 0.9
X_train = train_X[:int(split*train_X.shape[0]),:]
Y_train = train_Y[:int(split*train_Y.shape[0])]
X_test = train_X[int(split*train_X.shape[0]):,:]
Y_test = train_Y[int(split*train_Y.shape[0]):]
```

```python
print(X_test.shape , X_train.shape , Y_train.shape , Y_test.shape )
```

```python
ones  = np.ones((X_train.shape[0],1))
X_train = np.hstack((ones,X_train))
ones  = np.ones((X_test.shape[0],1))
X_test = np.hstack((ones,X_test))
#adding the column of X0
```

```python
print(X_test.shape , X_train.shape , Y_train.shape , Y_test.shape )
```

```python
def sigmoid(X):
    return 1/(1+np.exp(-X))
```

```python
def hypothesis(X,weights):
    return sigmoid(np.dot(X,weights))
```

```python
def loss(X,Y,weights):
    pred = hypothesis(X,weights)
    loss = np.mean(Y* np.log(pred) + (1-Y)*np.log(1-pred))
    return -1*loss
```

```python
def gradient(X,Y,weights):
    pred = hypothesis(X,weights)
    grad = np.dot(X.T,(Y-pred))
    return grad/X.shape[0]
```

```python
def gradient_descent(X,Y):
    weights = np.zeros((X.shape[1],))
    epochs = 100000
    lr_rate = 0.00008
    list_loss = []
     
    for i in range(epochs):
        grad = gradient(X,Y,weights)
        list_loss.append(loss(X,Y,weights))
        weights = weights + (lr_rate*grad)
    return weights,list_loss
```

```python
weights , list_loss = gradient_descent(X_train,Y_train)
```

# Model Testing

```python
pred = []

for i in range(X_test.shape[0]):
    y_hat = hypothesis(X_test[i],weights)
    print(y_hat , Y_test[i])
    if (y_hat > 0.5):
        pred.append(1)
    else:
        pred.append(0)
```

```python
for i in range(Y_test.shape[0]):
    print(pred[i] , '  ' , Y_test[i])
```

# Calculating Accuracy

```python
correct = 0
for i in range(Y_test.shape[0]):
    if(pred[i] == Y_test[i]):
        correct += 1
print(correct/Y_test.shape[0])
```
